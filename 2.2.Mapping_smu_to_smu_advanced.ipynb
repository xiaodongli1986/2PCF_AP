```python
def Rebin_2darray(A, n1=1,n2=1):
    numrow1A, numrow2A = len(A), len(A[0])
    numrow1B, numrow2B = len(A)/n1, len(A[0])/n2
    B = [[0 for row2 in range(numrow2B)] for row1 in range(numrow1B)]
    for row1 in range(numrow1A):
        for row2 in range(numrow2A):
            row1B, row2B = int(row1/n1+0.5), int(row2/n2+0.5)
            B[row1B][row2B] += A[row1][row2]
    return B

def Rebin_3darray(A, n1=1,n2=1,compute_rows=None):
    numrow1A, numrow2A = len(A), len(A[0])
    numrow1B, numrow2B = len(A)/n1, len(A[0])/n2
    numrow3 = len(A[0][0])
    B = [[[0 for row3 in range(numrow3)] for row2 in range(numrow2B)] for row1 in range(numrow1B)]
    for row1 in range(numrow1A):
        for row2 in range(numrow2A):
            row1B, row2B = int(row1/n1+0.5), int(row2/n2+0.5)
            if compute_rows == None:
                for row3 in range(numrow3):
                    B[row1B][row2B][row3] += A[row1][row2][row3]
            else:
                for row3 in compute_rows:
                    #print row1B, row2B, row3
                    B[row1B][row2B][row3] += A[row1][row2][row3]
    return B
Rebin_2darray([[1,2,3,4],[1.01,2,3,4],[1,2,3,4],[1,2.02,3,4]],2,2)
```
