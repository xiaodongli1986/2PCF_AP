```python
# sigma = s * cos theta = s * mu = s * (1-angle)
# pi = (s^2 - sigma^2) ^ 0.5

def sigmapi_from_sangle(s, angle):
    mu = 1-angle;
    sigma =  s*mu;
    pi = np.sqrt(s*s - sigma*sigma);
    return sigma, pi
def sangle_from_sigmapi(sigma, pi):
    s = np.sqrt(sigma*sigma+pi*pi)
    mu = pi/s
    angle = 1-mu
    return s, angle
```

```python
# Mapping from s-mu to sig-pi:

maxrow = 20
Xisigpi_2d_fromsmu = [[0 for row1 in range(maxrow)] for row2 in range(maxrow)]
for isig in range(maxrow):
    for ipi in range(maxrow):
        s, angle = sangle_from_sigmapi(isig+0.5,ipi+0.5)
        sindex = s - 0.5
        angleindex = angle / deltamu - 0.5
        #print isig, ipi, s, angle, sindex, angleindex
        x1, y1 = int(sindex), int(angleindex)
        x2, y2 = x1+1, y1+1
        xi_from_smu = LinearInterpolation_2d(x1, y1, x2, y2, \
                               Xismu_2d[x1][y1], Xismu_2d[x1][y2], \
                               Xismu_2d[x2][y1], Xismu_2d[x2][y2], \
                               sindex, angleindex)
        Xisigpi_2d_fromsmu[isig][ipi] = xi_from_smu
        
# Mapping from sig-pi to s-mu:
Xismu_2d_fromsigpi = [[0 for row1 in range(nummu)] for row2 in range(maxrow)]
for i_s in range(maxrow):
    for iangle in range(nummu):
        sig, pi = sigmapi_from_sangle(i_s+0.5, iangle*deltamu+deltamu*0.5)
        sigindex = sig - 0.5
        piindex = pi - 0.5
        x1, y1 = int(sigindex), int(piindex)
        #print x1, y1
        x1, y2 = x1+1, y1+1
        xi_from_sigpi = LinearInterpolation_2d(x1, y1, x2, y2, \
                               Xisigpi_2d[x1][y1], Xisigpi_2d[x1][y2], \
                               Xisigpi_2d[x2][y1], Xisigpi_2d[x2][y2], \
                               sigindex, piindex)
        Xismu_2d_fromsigpi[i_s][iangle] = xi_from_sigpi
```
