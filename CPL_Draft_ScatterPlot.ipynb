```python
### Read in data; filt data to redshift limits of LOWZ, CMASS
radecdata1 = np.loadtxt('/home/xiaodongli/SparseFilaments/data/input/DR12/galaxy_DR12v5_LOWZ_North_t1.txt')
radecdata2 = np.loadtxt('/home/xiaodongli/SparseFilaments/data/input/DR12/galaxy_DR12v5_CMASS_North_t1.txt')
xyzwdata1  = np.loadtxt('/home/xiaodongli/SparseFilaments/data/input/DR12/galaxy_DR12v5_LOWZ_North_t1.txt.xyzw')
xyzwdata2  = np.loadtxt('/home/xiaodongli/SparseFilaments/data/input/DR12/galaxy_DR12v5_CMASS_North_t1.txt.xyzw')
def filt_z(A, zlim):
    z1, z2 = zlim
    B = []
    for X in A:
        if z1 < X[2] < z2:
            B.append(X)
    return B
def randpart(A, randrat = 1.0):
    B = []
    for X in A:
        if random.uniform(0,1) < randrat:
            B.append([x for x in X])
    return B
radecdata1 = randpart(filt_z(radecdata1, [0, 0.43]))
radecdata2 = randpart(filt_z(radecdata2, [0.43, 1.0]))
X = Xfromdata(radecdata1, 2); plt.hist(X)
X = Xfromdata(radecdata2, 2); plt.hist(X)
```

```python
for plot_black in [False,True]:
    icosmo = 1
    if plot_black:
        cline = 'lightgreen'
        ctext = 'darkgreen'
        cmarker='w'
    else:
        ctext = cline = 'k'
        cmarker='k'
        

    fig = plt.figure(figsize=(24,9))
    for cosmo in [[0.26, -1, 0.0], [0.26, -1, -2]]:
        nowom, noww0, nowwa = cosmo
        radeclimits = [140, 170, 10, 13]

        def filt_data(A, limits):
            B = []
            xmin, xmax, ymin, ymax = limits
            for X in A:
                if xmin < X[0] < xmax and ymin < X[1] < ymax:
                    B.append([x for x in X])
            return B

        ### ra, dec limited sample
        B = filt_data(radecdata1, radeclimits) + filt_data(radecdata2, radeclimits)
        # Plot it
        X, Y = XYfromdata(B)
        if False:
            fig, ax = figax()
            ax.scatter(X, Y, s=0.1)

        ### ra, dec mean
        ramean = (radeclimits[0]+radeclimits[1])/2.
        decmean = (radeclimits[2]+radeclimits[3])/2.

        ### x, y, z of the ra, dec limited sample
        C = []
        for X in B:
            ra, dec, z = X[0], X[1], X[2]
            ra -= ramean; dec -= decmean
            r = comov_r_CPL(nowom, noww0, nowwa, 0.7, z)
            x, y, z = radecr_to_xyz(ra, dec, r)
            C.append([x, y, z])
        ### Plot it    
        X, Y, Z = XYZfromdata(C)
        if False:
            fig =  plt.figure(); ax = fig.add_subplot(111, projection = '3d')
            ax.scatter(X, Y, Z, s=0.1)


        ### Make final fan plot
        X, Y, Z = XYZfromdata(C)
        #fig, ax = figax(16,8.5); 
        ax = fig.add_subplot(1,2,icosmo)

        addts = [0.04, 0.025, 0.02, 0.018, 0.015, 0.014, 0.013]
        delta_ra = (radeclimits[1] - radeclimits[0]) / 2.0
        thetamin, thetamax = - delta_ra * np.pi/180., delta_ra * np.pi/180.
        for iz in range(len(Seven_Edges)):
            nowz = get_z_dft(Seven_Edges[iz])
            nowr = comov_r_CPL(nowom, noww0, nowwa, 0.7, nowz)

            nowring = [[cos(t)*nowr, sin(t)*nowr] for t in np.linspace(thetamin, thetamax, 100)]
            X0, Y0 = XYfromdata(nowring); ax.plot(X0, Y0, c=cline, lw=2)
            nowt = thetamax+addts[iz]
            meant = thetamax + sum(addts)/float(len(addts))
            ax.text(cos(nowt)*nowr, sin(nowt)*nowr, '$z=%.2f'%nowz+'$', fontsize=22, color=ctext, rotation = thetamax * 180/np.pi,
                    horizontalalignment='center', verticalalignment='center',)
            ax.text(cos(-nowt)*nowr, sin(-nowt)*nowr, '$r=%i'%nowr+'$', fontsize=20, color=ctext, rotation = thetamin * 180/np.pi,
                    horizontalalignment='center', verticalalignment='center',)

        rmax = comov_r_CPL(nowom, noww0, nowwa, 0.72, 0.72);
        Xfill = [cos(thetamax)*rmax*0.2, cos(thetamax)*rmax*1.05]
        Y1 = [sin(thetamax)*rmax*0.2, sin(thetamax)*rmax*1.05]
        Y2 = [-sin(thetamax)*rmax*0.2, -sin(thetamax)*rmax*1.05]
        if plot_black:
            ax.fill_between(Xfill, Y1, Y2, color='k' )
        else:
            ax.plot(Xfill,Y1, color='k')
            ax.plot(Xfill,Y2, color='k')

        X3 = np.linspace(0, comov_r_CPL(nowom, noww0, nowwa, 0.7, get_z_dft(Seven_Edges[6])), 21)
        Y3 = [0 for x in X3]
        ax.plot(X3, Y3, lw = 2, color=cline )
        X3 = [X3[row] for row in range(1,len(X3))]
        Y3 = [0 for x in X3]
        if plot_black:
            ax.plot(X3, Y3, marker='|', markersize=20, markeredgewidth=2, color=cline)

        ax.set_xlim(rmax*0.2,rmax); ax.set_ylim(-rmax/3.4, rmax/3.4);     
        ax.scatter(X,Y, marker='o', color = cmarker, linewidth=0, s = 2.5)
        ax.set_frame_on(False)
        ax.set_xticks([]); ax.set_yticks([]);
        if icosmo == 1:
            nowtext = '$\\rm \\Omega_m=%.2f'%nowom+'\\ \\Lambda CDM\\ Cosmology$'
        elif icosmo == 2:
            #nowtext = '$\\rm de\\ Sitter\\ cosmology\\ (cc\\ dominated)$'
            #nowtext = '$\\rm SCDM\\ cosmology\\ (DM\\ dominated)$'
            nowtext = '$\\rm \\Omega_m=%.2f'%nowom+',$ $w_0=-1,\\ w_a=-2$'
        ax.text(0.05, 0.05, nowtext, transform = ax.transAxes, fontsize = 30)

        icosmo += 1
    fig.tight_layout(w_pad=3)
    if plot_black:
        fig.savefig('./Draft/fig_fan.black.eps', format='eps')    
    else:
        fig.savefig('./Draft/fig_fan.eps', format='eps')    
```
