```python
execfile('/home/xiaodongli/software/pythonlib/stdA.py')
%pylab inline


om_dft, w_dft, h_dft = 0.26, -1, 0.7
assumed_cosmo = [ 
                  [om_dft,-1.5,0,h_dft],
                  [om_dft,-0.5,0,h_dft], 
                  [om_dft,-1.0,2,h_dft],
                  [om_dft,-1.0,-2,h_dft],
                  ]

#cosmo_names = [ '$\\rm \\Omega_m=%.2f'%omw[0]+'$, '+'$ w=%.1f'%omw[1]+'$' for omw in assumed_cosmo]
cosmo_names = []

for omw in assumed_cosmo:
    if omw[2] == 0:
        cosmo_names.append('$\\rm \\Omega_m=%.2f'%omw[0]+'$, '+'$ w=%.1f'%omw[1]+'$')
    else:
        cosmo_names.append('$\\rm \\Omega_m=%.2f'%omw[0]+'$, $w_0=%.1f'%omw[1]+',\\ w_a=%.1f'%omw[2]+'$')

def mus_relation_CPL(mu0,s0,om0,w0,om1,w1,wa1,z):
    alpha_perpen = comov_r_CPL(om1,w1,wa1,h_dft, z)/ comov_r(om0,w0,h_dft, z)
    alpha_parall = Hz(om0,w0,h_dft, z) / Hz_CPL(om1,w1,wa1,h_dft, z)
    factor = np.sqrt(  (alpha_parall*mu0)**2.0 + alpha_perpen**2.0 * (1.0-mu0**2.0))
    mu1 = alpha_parall * mu0 / factor
    s1 = s0 * factor
    return mu1, s1

fig=plt.figure(figsize=((len(assumed_cosmo))*4,(len(assumed_cosmo))*4))

### properties of plottings
ls0,ls1='--','-'; lw0,lw1=2,0.5; msize = 40; c0,c1 = 'b', 'r'
ngrids, ngridmu = 6, 13

plot_dense = True
ngrids_dense, ngridmu_dense = (ngrids-1)*5+1, (ngridmu-1)*5+1

red = 0.6

for icosmo in range(len(assumed_cosmo)):
        ### grid points of mu, s
        mu_list = np.linspace(0,1,ngridmu); 
        mu_list_dense = np.linspace(0,1,ngridmu_dense); 
        s_list = np.linspace(0,50,ngrids)
        s_list_dense = np.linspace(0,50,ngrids_dense)
        
        mus_list0, mus_list1, mus_list1_dense = [], [], []
        mu_list0, s_list0 = mu_list, s_list
        
        ### 2d mu, s list, in original and new cosmologies
        mus_2dlist1 = [[0 for i_s in range(ngrids)] for i_mu in range(ngridmu)]
        mus_2dlist1_dense = [[0 for i_s in range(ngrids_dense)] for i_mu in range(ngridmu_dense)]
        
        ax=fig.add_subplot(2,len(assumed_cosmo)/2,icosmo+1);
        om1,w1,wa1 = assumed_cosmo[icosmo][0], assumed_cosmo[icosmo][1], assumed_cosmo[icosmo][2]
        print 'mu, s distortion in cosmology ', om1, w1, wa1, ', redshift = ', red

        
        i_mu = -1
        if plot_dense:    
            for mu in mu_list_dense:
                i_mu +=1; i_s=-1
                for s in s_list_dense:
                    i_s += 1; mu1, s1 = mus_relation_CPL(mu,s,0.26,-1, om1,w1,wa1, red)
                    mus_list1_dense.append([mu1,s1]);
                    mus_2dlist1_dense[i_mu][i_s] = [mu1,s1]
        i_mu = -1
        if True:
            for mu in mu_list:
                i_mu +=1; i_s=-1
                for s in s_list:
                    i_s += 1; mu1, s1 = mus_relation_CPL(mu,s,0.26,-1, om1,w1,wa1, red)
                    mus_list0.append([mu,s]);
                    mus_list1.append([mu1,s1]);
                    #print i_mu, i_s, len(mus_2dlist1), len(mus_2dlist1[0])
                    mus_2dlist1[i_mu][i_s] = [mu1,s1]
        
        ### make plottin
        X0, Y0 = XYfromdata(mus_list0); X1, Y1 = XYfromdata(mus_list1)
        ### grid points, original cosmology
        #ax0.scatter(X0, Y0, marker='o',linewidths=0,s=msize)
        #ax2.scatter(X0, Y0, marker='o',linewidths=0,s=msize)
        ### grid points, new cosmology
        #ax1.scatter(X1, Y1, marker='*',s=msize,c='k')
        #ax2.scatter(X1, Y1, marker='*',s=msize,c='k')
        for imu in range(ngridmu):
            ### mu, s lines, for original cosmology
            nowmu_list0 = [mu_list0[imu] for row in range(ngrids)]
            nows_list0 = s_list0
            ax.plot(nowmu_list0, nows_list0, ls=ls0, lw=lw0, c=c0)
            ### mu, s lines, for new cosmology
        
        if plot_dense:
            for imu in range(ngridmu_dense):
                nowmu_list1_dense = [mus_2dlist1_dense[imu][i_s][0] for i_s in range(ngrids_dense)]
                nows_list1_dense = [mus_2dlist1_dense[imu][i_s][1] for i_s in range(ngrids_dense)]
                ### make plot
                ax.plot(nowmu_list1_dense, nows_list1_dense, ls=ls1, lw=lw1, c=c1)
        else:
            for imu in range(ngridmu):
                nowmu_list1 = [mus_2dlist1[imu][i_s][0] for i_s in range(ngrids)]
                nows_list1 = [mus_2dlist1[imu][i_s][1] for i_s in range(ngrids)]
                ### make plot
                ax.plot(nowmu_list1, nows_list1, ls=ls1, lw=lw0, c=c1)
        for i_s in range(ngrids):
            ### mu, s lines, for original cosmology
            nowmu_list0 = mu_list0
            nows_list0 = [s_list0[i_s] for row in range(ngridmu)]
            ax.plot(nowmu_list0, nows_list0, ls=ls0, lw=lw0, c=c0)
            ### mu, s lines, for new cosmology
        
        if plot_dense:
            for i_s in range(ngrids_dense):
                nowmu_list1_dense = [mus_2dlist1_dense[imu][i_s][0] for imu in range(ngridmu_dense)]
                nows_list1_dense = [mus_2dlist1_dense[imu][i_s][1] for imu in range(ngridmu_dense)]
                ### make plot
                ax.plot(nowmu_list1_dense, nows_list1_dense, ls=ls1, lw=lw1, c=c1)
        else:
            for i_s in range(ngrids):
                nowmu_list1 = [mus_2dlist1[imu][i_s][0] for imu in range(ngridmu)]
                nows_list1 = [mus_2dlist1[imu][i_s][1] for imu in range(ngridmu)]
                ### make plot
                ax.plot(nowmu_list1, nows_list1, ls=ls1, lw=lw1, c=c1)
            
        ### 
        if icosmo==0 or icosmo== 2:
            ax.set_ylabel('$s \\ \\rm[Mpc/h]$',fontsize=20);
        else:
            for xlabel_i in ax.get_yticklabels():
                xlabel_i.set_fontsize(0.0)
                xlabel_i.set_visible(False)
        
        if icosmo ==0 or icosmo ==1:
            for xlabel_i in ax.get_xticklabels():
                xlabel_i.set_fontsize(0.0)
                xlabel_i.set_visible(False)
        else:
            ax.set_xlabel('$\\mu$',fontsize=20);
                #if mod(k,2) == 0 or (i==3 and k==1):
                #    xlabel_i.set_visible(False)
        text = str(cosmo_names[icosmo])
        ax.set_xlim(0.0008,0.999);ax.set_ylim(0.05,65)
        ax.text(0.05,0.9, text, fontdict={'size': 24},
                    horizontalalignment='left',verticalalignment='center',transform = ax.transAxes);
fig.tight_layout();
fig.subplots_adjust(wspace = 0.05, hspace = 0.07);
plt.savefig('./Draft/fig_smu.eps',format='eps');
plt.show();
```
