```python
### SPH smoothed particle hydrodynamics paper will give yo;u the functional form of Spline Kernel. 
# Adopt n=30, for example, and use the variable h. Actually you can calculate the density and density
# gradiant at the same time using the derivative of the kernel. Monaghan ...

execfile('/home/xiaodongli/software/pythonlib/stdA.py')

import matplotlib.pyplot as plt
import numpy as np
import commands
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import os
import random
import scipy.integrate

#################################
# Useful Definiations
#################################

FIG_X_SIZE = 7
FIG_Y_SIZE = 5

CONST_C = 299792.458 #unit: km/s

PLOT_COLOR_ARRAY = ['k', 'b', 'g', 'r', 'c', 'm', 'y']
PLOT_STYLE_ARRAY = ['-',  '--', '-.' ]#,  'o', 'v', '+', 'x', 's', '*', 'D', ':']

NOW_PLOT_COLOR = 0
NOW_PLOT_STYLE = 0
NUM_PLOT_COLOR = len(PLOT_COLOR_ARRAY)
NUM_PLOT_STYLE = len(PLOT_STYLE_ARRAY)

def get_color_style():
    global NOW_PLOT_COLOR, NOW_PLOT_STYLE
    i = mod(NOW_PLOT_COLOR, NUM_PLOT_COLOR)
    j = mod(NOW_PLOT_COLOR, NUM_PLOT_STYLE)
    color_style = PLOT_COLOR_ARRAY[i]+PLOT_STYLE_ARRAY[j]
    NOW_PLOT_COLOR = NOW_PLOT_COLOR + 1
    NOW_PLOT_STYLE = NOW_PLOT_STYLE + 1
    return color_style

#################################
# get theta, phi according to x,y. 2D
#################################
def getangle(x, y):
    theta = np.arcsin(y/np.sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta
#################################
# get r, theta, phi given x,y,z. 3D
#################################
def getSC(x, y, z):
    r=np.sqrt(x**2+y**2+z**2)
    theta=getangle(z,np.sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi
#################################
# Find Root: f(x) == f0
#################################
def FindRoot( f, f0, xl, xr, tol = 1e-6, maxstep=100 ):
    fl = f(xl); fr = f(xr);
    if abs(fl-f0) < tol:
        return xl, fl-f0, 0, 'S'
    if abs(fr-f0) < tol:
        return xr, fr-f0, 0, 'S'
    if fl > f0 and fr > f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both larger than ", f0, "!"
        return 0, 0, 0, 'E' # E means error
    if fl < f0 and fr < f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both smaller than ", f0, "!"
        return 0, 0, 0, 'E'
    
    xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
    step = 0
    while abs(fm-f0) > tol and step <= maxstep:
        if (fm > f0 and fl > f0) or (fm < f0 and fl < f0):
            xl = xm; fl = fm
        else:
            xr = xm; fr = fm
        xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
        step +=1
    return xm, fm-f0, step, 'S' # S means success 

#################################
# Hubble parameter in unit of km/s/Mpc
#################################
def Hz(omegam, w, h, z):
    return 100*h*np.sqrt(omegam*(1.0+z)**3.0 + (1.0-omegam)*(1.0+z)**(3.0*(1+w)))

#################################
# comoving r in unit of Mpc/h
# in fact the result is independt on the input variable h
#################################
def comov_r(omegam, w, h, z):
    global CONST_C
    x, y = scipy.integrate.quad(lambda x: 1.0/Hz(omegam, w, h, x), 0, z)
    return CONST_C * x * h 
    
#################################
# invert of the comov_r function: 
# get z according to comoving_r 
# (in unit of Mpc/h)
#################################
def get_z(omegam, w, h, given_comov_r, zl=0.0, zr=5.0):
    z, deltar, num_step, flag = FindRoot(lambda z: comov_r(omegam, w, h, z), given_comov_r, zl, zr)
    return z
#################################
# unit transition: from km to Mpc/h
#################################
def km_to_Mpctoh(v, h):
    return v / (3.08567758e19) * h

#### 
# mean, variance (with error) of a list

def find_mean_var(n_list):
    n = len(n_list)
    n_mean = sum(n_list) / (n+0.0)
    n_var = sum([(n_list[row]-n_mean)**2.0 for row in range(n)]) / (n-1.0)
    return n_mean, n_var

def XYZfromdata(data, i=0,j=1,k=2):
    ndat = len(data)
    return [data[row][i] for row in range(ndat)], [data[row][j] for row in range(ndat)], \
        [data[row][k] for row in range(ndat)]

def ListXfromdata(data, ilist):
    ndat = len(data)
    return ([data[row][i] for row in range(ndat)] for i in ilist)

def Xsfromdata(data, ilist):
    return ([ data[row][ilist[i]] for row in range(len(data))] for i in range(len(ilist)))

def get_str_list(str1, div = [' ', ',', '\t', '\n']):
    strlist = []
    str2 = ''
    numberunderconstruct = False
    for i in range(len(str1)):
        duru = str1[i]
        strisdiv = False
        for sepstr in div:
            if duru == sepstr:
                strisdiv = True
                if numberunderconstruct == True:
                    strlist.append(str2); str2 = ''
                    numberunderconstruct = False
        if strisdiv == False:
            if numberunderconstruct == False:
                numberunderconstruct = True
                str2 = duru
            else:
                str2 += duru
    if str2 != '':
        strlist.append(str2);
    return strlist

##########################
## Conversion between numbers and characters/strings;
def str_to_numbers(str1,exitcode='#', do_float_conver=True):
    floatlist = []
    str2 = ''
    numberunderconstruct = False
    for i in range(len(str1)):
        duru = str1[i]
        if duru == exitcode:
            break
        elif duru == '\n' or duru == ' ' or duru == '\t' or duru == ',':
            if numberunderconstruct == True:
                floatlist.append(str2); str2 = ''
                numberunderconstruct = False
        else:
            if numberunderconstruct == False:
                numberunderconstruct = True
                str2 = duru
            else:
                str2 += duru
    if str2 != '':
        floatlist.append(str2);
    if do_float_conver:
        floatlist = [float(floatlist[row]) for row in range(len(floatlist))]
    return floatlist

STot = 4*(np.pi)*(180/np.pi)**2.0
```
